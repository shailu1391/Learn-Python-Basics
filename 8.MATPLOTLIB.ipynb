**MATPLOTLIB**

Matplotlib is a plotting library for Python language and it is a numerical extension of numpy.It provides an object oriented API for embedding plots into applications using general purpose GUI tooltiks like Tkinter, WXPython, Qt or GTK+

Other visualisaton platform is seaborn(for statistical analysis)

Pros of Matplotlib

1.Easy to get started for simple plots, 2D and 3D plots

2.Supports custom label and texts

3.Great control for every element in a figure

4.High quality output in many formats

5.Very customisable in general

```python
#compute x and y cordinates of a sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave")
plt.plot(x,y,'r')
```

```python
#use np.sin(x) and np.cos(x) for plotting graphs
```

```python
x=np.arange(0,4*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin,'g')
plt.title("Sinewave")
plt.subplot(2,1,2)
plt.plot(x,y_cos,'m--')
plt.title("Coswave")
```

**BOX PLOTS**

USED FOR PERCENTILE GRAPHS

```python
data =[np.random.normal(0,std,100) for std in range(1,4)]

#rectangular box plot

plt.boxplot(data,vert=True,patch_artist=True); #note the semicolon
```

**PIE CHARTS**

```python
labels="Chocolate","Cookie","Shake","Cakes"
sizes=[20,40,60,90]
colors=["green","lightblue","lightcoral","magenta"]
explode=(0.4,0,0.2,0) #displacing first slice

#plotting

plt.pie(sizes,explode=explode,colors=colors,labels=labels,autopct='%1.1f%%',shadow=False)
##autopct indicates format of values 1.1f is floating format
plt.axis('equal')
```
