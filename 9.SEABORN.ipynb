**SEABORN**

It is the backbone of exploratory data analysis

Some background on Datasets:

Whenever there is a data set, classify the data in terms of its features(eg. f1,f2,f3)and also as dependable and independent features. Seaborn is good for univariate and bivariate data analysis. But for multivariate data analysis following distribution plots can be used.

**Distribution plots**

1.displot

2.jointplot

3.pairplot


**Categorical plots**

1.boxplot

2.violinplot

3.countplot

4.barplot

Practice problem on IRIS data set

```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.head() #TIP IS THE DEPENDENT FEATURE
```

**CORRELATION WITH HEATMAPS**

A Correlation heatmap uses colored cells in a 2D matrix to show relation between 2 discreet elements or event types. It is very important in feature selection.

```python
df.corr()
```

```python
df.dtypes #heatmaps and correlation can only be done for float and int variables
```

```python
sns.heatmap(df.corr())
```

**JOINT PLOT**

FOR UNIVARIATE AND BIVARIATE ANALYSIS

```python
#Doing bivariate analysis using tip and total bill as two features
#jointplot also gives histogram on the outline
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind='hex')
```

```python
#using kind as 'reg' gives a regression line (the best fit line)
#bell curve indicates probability density function
```

```python
sns.jointplot(x="tip",y="total_bill",data=df,kind='reg')
```

**PAIR PLOTS**

Pair plots are also called scatter plots where one variable data in a row is matched with another variables value.

It is used where there are more than two independent features where it will do permutation and combination of all features.(integer or floating values)

```python
sns.pairplot(df)
```

```python
#hue parameter is used to classify pairplots using other variables like sex/smoker
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts() #to know the count of number of smokers
```

**DIS PLOTS**

Dis plots help in checking the column feature

```python
sns.displot(df['tip'])
```

```python
sns.distplot(df['size'],kde=True,bins=20)
#kde is kernal density estimation indicated by the curved line
```

**CATEGORICAL PLOTS**

Analysis using categorical data points (those that are not float or integer type)

```python
#COUNTPLOT
#Show the counts of observations in each categorical bin using bars.
#give either x or y value not both, as axis always gives count

sns.countplot(x='sex',data=df)
```

```python
sns.countplot(y='smoker',data=df)
```

```python
sns.countplot(x='day',data=df)
```

```python
#Bar plot also does the same thing but both x and y values should be given
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
#box plot gives percentile distribution
```

```python
sns.boxplot(x='sex',y='total_bill',data=df,hue='smoker',palette='rainbow')
#palette is optional
```

```python
sns.boxplot(data=df,orient='v')
#considers all features by default
#orient is vertical or horizontal orientation
```

**Violin Plot**

violin plot helps in visualisation of data in terms of both kernal density distribution and the box plot

```python
sns.violinplot(x='total_bill',y='day',data=df)
```
